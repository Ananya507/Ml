```python
import pandas as pd
import os
os.getcwd()
```

```python
Data=pd.read_csv('titanic.csv')
Data
```

```python
Data.isna().sum()
```

```python
Data=Data.dropna()
Data
```

```python
Data
```

```python
data1=Data.drop('Name',axis=1)
data2=data1.drop('Ticket',axis=1)
data3=data2.drop('PassengerId',axis=1)
data3
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data3=data3.apply(le.fit_transform)
```

```python
data3
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier(criterion='entropy')
```

```python
x=data3.drop(columns=['Survived'])
y=data3['Survived']
```

```python
model.fit(x,y)
```

```python
model.predict(x)
```

```python
model.score(x,y)
```

```python
from sklearn import tree
tree.plot_tree(model,feature_names=x.columns);
```
